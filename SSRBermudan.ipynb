We start off by implementing a Discretization Scheme based on the Black-Scholes model.

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 0.001
sigma = 0.01
start_price = 5

def StockpriceGBM(T,S0,mu,sigma):
  WienerDifference = np.random.normal(0,T)
  return S0*np.exp((mu - 0.5*sigma**2)*T + sigma*WienerDifference)

def discretizationScheme(times,S0,mu,sigma):
  stockPrices = [S0]
  deltas = [0]

  for time in times:
    deltas.append(time - deltas[-1])
    Scurrent = stockPrices[-1]
    Snew = StockpriceGBM(deltas[-1],Scurrent,mu,sigma)
    stockPrices.append(Snew)

  return stockPrices
```

```python
discretizationScheme([1,2,3,4],start_price,mu,sigma)
```

```python
plt.plot(discretizationScheme(range(100),start_price,mu,sigma))
```
